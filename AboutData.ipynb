```python
import pandas as pd
```

```python
df = pd.read_csv('household_power_consumption.txt',
                  sep=';',           # semicolon separator
                  na_values='?',     # treat ? as missing
                  low_memory=False,
                  parse_dates={'datetime': ['Date', 'Time']},  # combine date+time
                  dayfirst=True)     # date format is DD/MM/YYYY
```
